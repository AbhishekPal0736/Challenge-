Logistics Regression classification problem

```python
import pandas as pd
import seaborn as sn
```

Read data

```python
df=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv')
```

Using some types of function

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

Define Y and X variables

```python
y=df['diabetes']
X=df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
sn.regplot(x='glucose',y='diabetes',data=df)
sn.regplot(x='glucose',y='diabetes',data=df,logistic=True)
```

Select model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

prediction

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

Find the accuracy of model

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
