Ordinal Encoding

```python
df=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Pokemon.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.drop(columns=["Type 2"],inplace=True)
```

```python
ord_data=[["Grass","Fire",'Normal', 'Fairy', 'Dragon', 'Ice', 'Electric', 'Ghost', 'Dark', 'Bug', 'Water', 'Flying', 'Poison', 'Steel', 'Rock', 'Fighting', 'Psychic', 'Ground']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(df[["Type 1"]])
```

```python
df["type 1_en"]=oe.transform(df[["Type 1"]])
```
